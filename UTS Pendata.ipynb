Gambar ini menunjukkan keseluruhan alur pengolahan data (workflow). #### 2. Missing Value
image.png
Tabel statistik awal menunjukkan adanya data hilang (missing values) yang signifikan. Pada tahap ini, saya melakukan imputasi data hilang menggunakan node Missing Value dengan metode Median untuk fitur angka agar hasil lebih stabil terhadap pencilan, dan Most Frequent Value untuk fitur teks (seperti tekstur tanah). Langkah ini memastikan data lengkap sebelum diproses lebih lanjut.

#### 3. Normalizer
image.png
image.png
KNN adalah algoritme berbasis jarak, sehingga fitur dengan rentang nilai besar (misalnya P-Tersedia) dapat mendominasi fitur dengan rentang nilai kecil (misalnya pH). Saya menggunakan node Normalizer dengan metode Min-Max untuk menyeragamkan skala semua fitur kimia tanah menjadi rentang 0 hingga 1, memastikan setiap fitur memiliki bobot yang adil.

#### 4. Confution Matrix
image.png
Confusion Matrix menyajikan detail performa prediksi model KNN terhadap data uji. Tabel ini menunjukkan jumlah total prediksi yang tepat untuk kelas 'Subur' dan 'Tidak Subur' serta detail kesalahan prediksinya. Ini memberikan gambaran yang lebih dalam mengenai kualitas model daripada sekadar angka akurasi.
